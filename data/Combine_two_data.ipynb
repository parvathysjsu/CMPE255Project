### Remove rows with 'M' values in max_temperature	min_temperature	avg_temperature	departure_temperature	hdd	cdd	percipitation columns

```python
def remove_rows_with_M_Values(weatherdata):
    df = weatherdata[weatherdata.max_temperature == 'M']
    print("remove: ", df.index)
    weatherdata = weatherdata.drop(df.index)
    df = weatherdata[weatherdata.min_temperature == 'M']
    print("remove: ", df.index)
    weatherdata = weatherdata.drop(df.index)
    df = weatherdata[weatherdata.avg_temperature == 'M']
    print("remove: ", df.index)
    weatherdata = weatherdata.drop(df.index)
    df = weatherdata[weatherdata.departure_temperature == 'M']
    print("remove: ", df.index)
    weatherdata = weatherdata.drop(df.index)
    df = weatherdata[weatherdata.hdd == 'M']
    print("remove: ", df.index)
    weatherdata = weatherdata.drop(df.index)
    df = weatherdata[weatherdata.cdd == 'M']
    print("remove: ", df.index)
    weatherdata = weatherdata.drop(df.index)
    df = weatherdata[weatherdata.percipitation == 'M']
    print("remove: ", df.index)
    weatherdata = weatherdata.drop(df.index)
    return weatherdata
weatherdata = remove_rows_with_M_Values(weatherdata)
print("Current weather Shape", weatherdata.shape)
```

### Add a new attribute "stationCoord" to the weather dataset

```python
def addStationCoord(weather):
    tmpList = []
    for station in weather['name']:
        tmpList.append(stationCoords[station])
    weather['stationCoord'] = tmpList
    return weather

weatherdata = addStationCoord(weatherdata)
print(weatherdata.shape)
weatherdata.head(5)
```

### Update Time to time object

```python
from datetime import datetime

def change_string_to_time_Object(weatherdata):
    tmpList = []
    for time in weatherdata['date']:
        datetime_object = datetime.strptime(time, '%Y-%m-%d')
        tmpList.append(datetime_object)
    weatherdata = weatherdata.drop(['date'], axis=1)
    weatherdata['date'] = tmpList
    return weatherdata

weatherdata = change_string_to_time_Object(weatherdata)
weatherdata.head()
```

### Create a new attriubte "geometry" in geo 

```python
#make make a centroid list
geo['centroid'] = geo['geometry'].centroid
print(geo['centroid'])
```

```python
#geoCoord: first coordinate 
#coord: second coordinate (String input)
def calculateDistance(geoCoord, coord):
    return (float(geoCoord[0])-float(coord[0]))**2 + (float(geoCoord[1])-float(coord[1]))**2

#wildfire's centroid coodinate input
#return the closestStation's coordinate
def findClosestStation(coord):
    minDistance = 10**10
    for station in stationCoords.items():
        distance = calculateDistance(coord, station[1])
        if(distance < minDistance):
            closestStation = station[1]
            minDistance = distance
    return closestStation
```

### Create a new attribute name "nearest Station"
#### It stores the nearest station's coordinate in longtitude, latitude format

```python
i = 0
nearestStationList = []
for i in range(len(geo['centroid'])):
    closestStation = findClosestStation(geo['centroid'][i].coords[0])
    nearestStationList.append(closestStation)
geo['nearestStation'] = nearestStationList
geo['nearestStation']
```

### update time data

```python
from datetime import datetime

def change_string_to_time_Object(geo):
    length = len(geo)
    startList = []
    endList=[]
    for i in range(length):
        datetime_object = datetime.strptime(geo['STARTDATE'][i][0:10], '%Y-%m-%d')
        startList.append(datetime_object)
        datetime_object = datetime.strptime(geo['PERIMDATE'][i][0:10], '%Y-%m-%d')
        endList.append(datetime_object)
    geo = geo.drop(['STARTDATE', 'PERIMDATE'], axis=1)
    geo['start'] = startList
    geo['end'] = endList
    return geo

geo = change_string_to_time_Object(geo)
```

```python
##input 
def find_same_location_row(coord, weatherdata):
    return weatherdata.loc[weatherdata['stationCoord'] == coord]

def checkDate(rows, startDate, endDate):
    startIndex = -1
    endIndex = -1
    startExist = False
    endExist = False
    index = 0
    for time in weatherdata['date']:
        if(time.date() == startDate.date()):
            startExist = True
            startIndex = index
        if(time.date() == endDate.date()):
            endExist = True
            endIndex = index
        index = index + 1
    return startExist and endExist, startIndex, endIndex

def findAverageValue(rows, startDateIndex, endDateIndex):
    maxTmpSum, minTmpSum, avgTmpSum, avgPrecSum, avgHddSum, avgCddSum = 0, 0, 0, 0, 0, 0
    size = abs(endDateIndex - startDateIndex) + 1   ### geo row 154 has reverse start date and end date
    for row in rows.itertuples():
#         print(row)
        maxTmpSum += float(row.max_temperature)
        minTmpSum += float(row.min_temperature)
        avgTmpSum += float(row.avg_temperature)
        avgPrecSum += float(row.percipitation)
        avgHddSum += float(row.hdd)
        avgCddSum += float(row.cdd)
    return maxTmpSum/size, minTmpSum/size, avgTmpSum/size, avgPrecSum/size, avgHddSum/size, avgCddSum/size

def add_new_attribute(geo):
    maxTmpList = []
    minTmpList = []
    avgTmpList = []
    avgPrecList = []
    avghddList = []
    avgcddList =[]
    n = len(geo['nearestStation'])
    print(n)
    i = 0
    for row in geo.itertuples():
        print(i)
        rows = find_same_location_row(row.nearestStation, weatherdata)
        checkDateExist, startDateIndex, endDateIndex = checkDate(rows, row.start, row.end)
        if(checkDateExist):
            maxTmp, minTmp, avgTmp, avgPrec, avgHdd, avgCdd = findAverageValue(rows, startDateIndex, endDateIndex)
            #get the avg value and add it to the array
        else:
            #cannot find date, thusinitializae negative -999
            maxTmp, minTmp, avgTmp, avgPrec, avgHdd, avgCdd = -999, -999, -999, -999, -999, -999
        #add it to the array
        maxTmpList.append(maxTmp)
        minTmpList.append(minTmp)
        avgTmpList.append(avgTmp)
        avgPrecList.append(avgPrec)
        avghddList.append(avgHdd)
        avgcddList.append(avgCdd)
        i += 1
    geo['maxTmp'] = maxTmpList
    geo['minTmp'] = minTmpList
    geo['avgTmp'] = avgTmpList
    geo['avgPrec'] = avgPrecList
    geo['avgHdd'] = avghddList
    geo['avgCdd'] = avgcddList
    return geo
```
